```python
m, n, r = 100, 100, 50
x = Domain([0, 1], m)
mu = Domain([0, 1], n)
```

```python
def save_im(res, name):
    print(res.min(), res.max())
    fig, ax = plt.subplots()
    cs = plt.imshow(res, interpolation="nearest", origin="upper")
    plt.colorbar(cs)
    a = np.round(cmap(res/2+0.5)*256, decimals=0)
    img = np.array(a-1, dtype=np.uint8)
    from PIL import Image
    im = Image.fromarray(img)
    im.save(name+'.png')
    plt.close()
```

```python
u_hs = Heaviside()

X1 = u_hs.half_wave_odd(x(), mu())
save_im(X1, "C:/Users/florianma/Dropbox/Kol-N-width-Oslo-Ulm/Plots/heaviside.png")
X2 = u_hs(x(), mu())
save_im(X2, "C:/Users/florianma/Dropbox/Kol-N-width-Oslo-Ulm/Plots/heaviside_symm.png")
svd_basis1 = SVD(X1)
svd_basis2 = SVD(X2)
```

```python
for U in [svd_basis1.U, svd_basis2.U]:
    r = 6
    fig, ax1 = plt.subplots()
    for i in range(r):
        plt.plot(x(), U[:, i], "o", ms=1, color=cmap(i/(r-1)), label="mode {:.0f}".format(i))
    plt.xlim([0, 1])
    plt.grid(which="both")
    plt.show()
```

```python
fig, ax1 = plt.subplots()
for S in [svd_basis1.S, svd_basis2.S]:
    S = S/S[0].copy()  # normalize singular values to make for a better comparison!

    ax1.plot(np.arange(1, len(S)+1), S, marker=".", ms=1, lw=0)
ax1.set_xlabel("order")
ax1.set_ylabel("singular value")
ax1.set_yscale('log')
ax1.set_xlim([0, len(S)])
ax1.set_ylim([1e-3, 10])
ax1.grid(which="both")
plt.tight_layout()
plt.show()
```
