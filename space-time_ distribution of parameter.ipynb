```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.heaviside(x, 0.5)
```

```python
print("f(-0.1) =", f(-0.1))
print("f(0.0) =", f(0.0))
print("f(0.1) =", f(0.1))
```

```python
a, b, c = 10, 10, 10
```

```python
x = np.linspace(0, 1, a, endpoint=False)
v = 0.5
t = 1

plt.plot(x, f(x-v*t), "b.")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

```python
x = np.linspace(0, 1, a, endpoint=False)
v = 1
t = 0.5

plt.plot(x, f(x-v*t), "b.")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

```python
x = np.linspace(0, 1, a, endpoint=False)
v = 1
t = np.linspace(0, 1, c, endpoint=False)
mu = v*t
xx, mus = np.meshgrid(x, mu, indexing="ij")

A = f(xx-mus)
plt.imshow(A, origin="upper")
```

```python
plt.hist(mus.ravel(), range=[-1/a/2, 1+1/a/2], bins=a+1)
plt.show()
```

```python
x = np.linspace(0, 1, a, endpoint=False)
v = np.linspace(0, 1, b, endpoint=False)
t = 1
mu = v*t
xx, mus = np.meshgrid(x, mu, indexing="ij")

A = f(xx-mus)
print(A.shape)
plt.imshow(A)
```

```python
plt.hist(mus.ravel(), range=[-1/a/2, 1+1/a/2], bins=a+1)
plt.show()
```

```python
x = np.linspace(0, 1, a, endpoint=False)
v = np.linspace(0, 1, b, endpoint=False)
t = np.linspace(0, 1, b, endpoint=False)
xx, vv, tt = np.meshgrid(x, v, t)

A = f(xx-vv*tt)
space_time_snapshot_matrix = A.reshape(a*b, c)
plt.imshow(space_time_snapshot_matrix)
```

```python
mus = vv*tt
plt.hist(mus.ravel(), bins=50)
plt.show()
```
